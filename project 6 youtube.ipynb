## 2. Find out the top 10 rated videos

```python
d = data.groupby(['no_of_ratings']).max()
```

```python
e = d.sort_values('no_of_ratings',ascending = False)
```

```python
e = e.drop(columns = ['published','length','views','no_comment'])
```

```python
print('Top 10 rated videos\n',e.head(10))
```
